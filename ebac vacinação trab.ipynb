Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
import pandas as pd
import numpy as np

# Dados
data = {
    'Matemática': np.random.randint(0, 25, 100),
    'Português': np.random.randint(0, 19, 100),
    'Geografia': np.random.randint(0, 9, 100),
    'Inglês': np.random.randint(0, 9, 100),
    'História': np.random.randint(0, 9, 100),
    'Física': np.random.randint(0, 13, 100),
    'Química': np.random.randint(0, 13, 100)
}

# Criando DataFrame
df = pd.DataFrame(data)

# Tabela de número de questões por assunto
num_questoes = pd.Series({'Matemática': 24, 'Português': 18, 'Geografia': 8, 'Inglês': 8, 'História': 8, 'Física': 12, 'Química': 12})

# 1. Calcule o percentual de acerto dos alunos por assunto.
percentual_acerto = (df / num_questoes) * 100

# 2. Calcule o total de acertos de cada aluno.
total_acertos_por_aluno = df.sum(axis=1)

# 3. Calcule o percentual geral de cada aluno.
percentual_geral_por_aluno = (total_acertos_por_aluno / (num_questoes.sum())) * 100

# 4. Quantos alunos tiveram nota maior que 45?
alunos_acima_de_45 = percentual_geral_por_aluno[percentual_geral_por_aluno > 45].count()

# Exibindo resultados
print("1. Percentual de acerto por assunto:")
print(percentual_acerto)

print("\n2. Total de acertos por aluno:")
print(total_acertos_por_aluno)

print("\n3. Percentual geral por aluno:")
print(percentual_geral_por_aluno)

print("\n4. Número de alunos com nota maior que 45:", alunos_acima_de_45)
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# 1. Monte um novo DataFrame com os alunos desqualificados em Matemática, Física ou Química.
alunos_desqualificados = df[(df['Matemática'] == 0) | (df['Física'] == 0) | (df['Química'] == 0)]

# 2. Quantos são esses alunos?
num_alunos_desqualificados = alunos_desqualificados.shape[0]

# 3. Qual a média desses alunos em História e Geografia?
media_desqualificados = alunos_desqualificados[['História', 'Geografia']].mean()

# 4. Monte um DataFrame com os alunos que passaram para a segunda fase.
alunos_passaram_segunda_fase = df[(df['Matemática'] > 0) & (df['Física'] > 0) & (df['Química'] > 0)]

# Exibindo os resultados
print("1. Alunos desqualificados:")
print(alunos_desqualificados)

print("\n2. Número de alunos desqualificados:", num_alunos_desqualificados)

print("\n3. Média dos alunos desqualificados em História e Geografia:")
print(media_desqualificados)

print("\n4. Alunos que passaram para a segunda fase:")
print(alunos_passaram_segunda_fase)
````
